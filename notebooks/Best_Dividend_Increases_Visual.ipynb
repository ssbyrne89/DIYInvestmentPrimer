```python
dollars = pd.read_csv('the20_companies_w_best_YoY_divs_paid_increases.csv')
dollars
```

```python
pcts = pd.read_csv('the20_companies_w_best_YoY_divs_paid_pct_changes.csv')
pcts
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    name="Avg. Dividend Paid per Share",
    x=dollars.columns[1:], y=dollars.mean()[1:]
))
fig.update_xaxes(showgrid=True)
fig.update_layout(
    title="Avg. Dividend Paid per Share Over the Last 20 Years",
    xaxis=dict(
        title='Company'
    ),
    yaxis=dict(
        title='Avg Dollar Amount Paid per Share'
    )
)
py.plot(figure_or_data=fig, filename="Avg_Div_Paid_per_Share", auto_open=True)
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    name="Avg. YoY Pct Change in Dividends Paid per Share",
    x=pcts.columns[1:], y=pcts.mean()[1:]
))
fig.update_xaxes(showgrid=True)
fig.update_layout(
    title="Avg. Pct Change in Dividends Paid per Share",
    xaxis=dict(
        title='Company'
    ),
    yaxis=dict(
        title='Avg Pct Change in Dividends Paid per Share'
    )
)
py.plot(figure_or_data=fig, filename="Avg_YoY_Pct_Change_in_Divs_Paid_per_Share", auto_open=True)
fig.show()
```

```python
for col in dollars.columns[1:]:
  fig = go.Figure()

  fig.add_trace(go.Bar(
      name="Dollar Amount of Dividend Paid per Share",
      x=dollars['year'], y=dollars[col]
  ))
  fig.add_trace(go.Scatter(
      name="Percent Change from previous year",
      mode="markers+lines", x=pcts['year'], y=pcts[col]
  ))
  fig.update_xaxes(showgrid=True)
  fig.update_layout(title_text=col,
                   xaxis=dict(
        title='Year'
    ),
                    yaxis=dict(
        title='Percent Change AND Dollars'
    )
    )
  py.plot(figure_or_data=fig, filename=f"{col}_YoY_div_paid_per_share_w_pct_change", auto_open=True)
  fig.show()
```
